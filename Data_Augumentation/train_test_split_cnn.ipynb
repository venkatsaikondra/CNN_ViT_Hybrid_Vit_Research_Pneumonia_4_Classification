```python
import os
import shutil
import random
from tqdm import tqdm

def split_cnn_data(source_root, target_root, train_ratio=0.8, val_ratio=0.1, test_ratio=0.1):
    """
    Splits the augmented data into Train, Val, and Test folders while
    maintaining class balance.
    """
    classes = ['Covid-19', 'Normal', 'Pneumonia-Bacterial', 'Pneumonia-Viral']
    phases = ['train', 'val', 'test']

    # Create the directory structure
    for phase in phases:
        for cls in classes:
            os.makedirs(os.path.join(target_root, phase, cls), exist_ok=True)

    print(f"📂 Starting data split from: {source_root}")

    for cls in classes:
        source_dir = os.path.join(source_root, cls)
        images = [f for f in os.listdir(source_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

        # Shuffle to ensure randomness
        random.shuffle(images)

        # Calculate split points
        total = len(images)
        train_end = int(total * train_ratio)
        val_end = train_end + int(total * val_ratio)

        # Assign images to phases
        split_dict = {
            'train': images[:train_end],
            'val': images[train_end:val_end],
            'test': images[val_end:]
        }

        print(f"⚖️ {cls}: Total={total} | Train={len(split_dict['train'])} | Val={len(split_dict['val'])} | Test={len(split_dict['test'])}")

        # Copy files
        for phase in phases:
            for img_name in tqdm(split_dict[phase], desc=f"Copying {cls} ({phase})", leave=False):
                shutil.copy(
                    os.path.join(source_dir, img_name),
                    os.path.join(target_root, phase, cls, img_name)
                )

    print(f"\n✨ Split Complete! Data is ready in: {target_root}")

# --- PATHS ---
SOURCE = "/Users/kpvarma/PycharmProjects/CNN_ViT_Hybrid_Vit_Research_Pneumonia_4_Classification/CNN_Augmented_Data"
DESTINATION = "/Users/kpvarma/PycharmProjects/CNN_ViT_Hybrid_Vit_Research_Pneumonia_4_Classification/Train_Test_Split_CNN_Data"

split_cnn_data(SOURCE, DESTINATION)
```
