```python
# Define your source and destination paths
input_folder = '/Users/kpvarma/PycharmProjects/CNN_ViT_Hybrid_Vit_Research_Pneumonia_4_Classification/Vit_Augumented_Data'
output_folder = '/Users/kpvarma/PycharmProjects/CNN_ViT_Hybrid_Vit_Research_Pneumonia_4_Classification/Train_Test_Split_Vit_Data'

# Ensure the input folder exists before splitting
if os.path.exists(input_folder):
    # Split with a ratio of (train, val, test)
    # 0.8 training, 0.1 validation, 0.1 testing
    splitfolders.ratio(input_folder,
                       output=output_folder,
                       seed=42,
                       ratio=(.8, .1, .1),
                       group_prefix=None,
                       move=False) # 'move=False' keeps your augmented data intact

    print(f"Successfully created: {output_folder}")

    # Verify the structure
    for root, dirs, files in os.walk(output_folder):
        level = root.replace(output_folder, '').count(os.sep)
        indent = ' ' * 4 * (level)
        if level <= 2: # Show top 2 levels
            print(f"{indent}{os.path.basename(root)}/ - ({len(files)} images)")
else:
    print("Source directory not found. Please check the path.")
```
